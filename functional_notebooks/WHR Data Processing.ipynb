```python
# no family, 

data_2015 = data_2015[['Country', 'Happiness Rank', 'Happiness Score','Economy (GDP per Capita)',# 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Generosity', 'Trust (Government Corruption)']]
data_2015['Year'] = [2015 - 1] * len(data_2015)


data_2016 = data_2016[['Country', 'Happiness Rank', 'Happiness Score','Economy (GDP per Capita)',# 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Generosity', 'Trust (Government Corruption)']]
data_2016['Year'] = [2016 - 1] * len(data_2016)


data_2017 = data_2017[['Country', 'Happiness.Rank', 'Happiness.Score','Economy..GDP.per.Capita.',# 'Family',
       'Health..Life.Expectancy.', 'Freedom', 'Generosity', 'Trust..Government.Corruption.']]
data_2017['Year'] = [2017 - 1] * len(data_2017)


data_2018 = data_2018[['Country or region', 'Overall rank', 'Score','GDP per capita', 
       'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
data_2018['Year'] = [2018 - 1] * len(data_2018)


data_2019 = data_2019[['Country or region', 'Overall rank', 'Score','GDP per capita', 
       'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
data_2019['Year'] = [2019 - 1] * len(data_2019)


data_2020 = data_2020[['Country or region', 'Overall rank', 'Score','GDP per capita', 
       'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
data_2020['Year'] = [2020 - 1] * len(data_2020)


data_2021['Overall rank'] = np.arange(1, len(data_2021)+1)
data_2021 = data_2021[['Country name', 'Overall rank', 'Ladder score','Logged GDP per capita', 
       'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
data_2021['Year'] = [2021 - 1] * len(data_2021)
```

```python
columns = ['Country', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 
           'Freedom', 'Generosity', 'Perceptions of corruption', 'Year']

data_2015.columns = columns
data_2016.columns = columns
data_2017.columns = columns
data_2018.columns = columns
data_2019.columns = columns
data_2020.columns = columns
data_2021.columns = columns
```

```python
data = data_2015.append(data_2016, ignore_index=True)
data = data.append(data_2017, ignore_index=True)
data = data.append(data_2018, ignore_index=True)
data = data.append(data_2019, ignore_index=True)
data = data.append(data_2020, ignore_index=True)
data = data.append(data_2021, ignore_index=True)
```

```python
import plotly.offline as py
import plotly.graph_objs as go
import geopandas as gpd
import pycountry

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
        except:
            CODE.append('Unknown')
    return CODE

data.replace(['Russia'], 'Russian Federation', inplace=True)
data['Country Code'] = alpha3code(data['Country'].str.strip())
```
