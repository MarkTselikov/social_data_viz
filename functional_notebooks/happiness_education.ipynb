```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
whd = pd.read_csv("data/whr_data.csv")
edudata = pd.read_csv("data/edu_d.csv")
```

```python
whd.head()
```

```python
whd = whd.rename(columns = {'Country Code': 'Country_code'})
```

```python
whd.info()
```

```python
edudata = edudata.drop(['Time Code'],axis =1)
edudata.head()
```

```python
edudata.columns
```

```python
edudata.columns = ['Year','Country','Country_code','Trained teachers in upper secondary education (% of total teachers)','Trained teachers in upper secondary education, female (% of female teachers)','Trained teachers in primary education, female (% of female teachers)','Trained teachers in primary education (% of total teachers)','Literacy rate, youth (ages 15-24), gender parity index (GPI)','School enrollment, primary (gross), gender parity index (GPI)','School enrollment, secondary (gross), gender parity index (GPI)']
```

```python
edudata.info()
```

```python
# Year and country contained NaN
edudata = edudata[~edudata['Year'].isnull()]
edudata = edudata[~edudata['Country'].isnull()]
edudata = edudata[~edudata['Country'].isnull()]
edudata.Year = edudata.Year.astype(int)
```

```python
wedu = pd.merge(whd, edudata, how = 'inner', on = ['Country_code','Year'])
```

```python
wedu.head()
```

```python
wedu = wedu.rename(columns = {'Country_x': 'Country'})
wedu = wedu.drop(['Country_y'],axis =1)
wedu = wedu.drop(['Unnamed: 0'],axis =1)
```
