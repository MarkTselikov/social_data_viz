```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline 
```

```python
unemp_data = pd.read_csv('data/youth_unemployment.csv')
unemp_data.head()
```

```python
crime_data = pd.read_csv('data/crime_homicide.csv')
crime_data.head()
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(121)
sns.heatmap(unemp_data.isna())

plt.subplot(122)
sns.heatmap(crime_data.isna())
```

```python
unemp_data = pd.melt(unemp_data, id_vars=['Country Name', 'Country Code', 'Indicator Name'], 
                     value_vars=['2000', '2001', '2002', '2003', '2004',
                   '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                   '2014', '2015', '2016', '2017', '2018', '2019', '2020'])

unemp_data.columns = ['Country Name', 'Country Code', 'Indicator Name', 'Year', 'Unemployment Score']
unemp_data['Year'] = pd.to_numeric(unemp_data['Year'], errors='coerce')
unemp_data.head()
```

```python
crime_data = pd.melt(crime_data, id_vars=['Country Name', 'Country Code', 'Indicator Name'], 
                     value_vars=['2000', '2001', '2002', '2003', '2004',
                   '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                   '2014', '2015', '2016', '2017', '2018', '2019', '2020'])

crime_data.columns = ['Country Name', 'Country Code', 'Indicator Name', 'Year', 'Crime (Homicide) Score']
crime_data['Year'] = pd.to_numeric(crime_data['Year'], errors='coerce')
crime_data.head()
```

```python
data = pd.merge(crime_data, unemp_data, on=['Country Code', 'Year'])
data.head()
```

```python
data = data.drop('Indicator Name_x', axis=1)
data = data.drop('Indicator Name_y', axis=1)
data = data.drop('Country Name_y', axis=1)

data.columns = ['Country Name', 'Country Code', 'Year', 'Crime (Homicide) Score', 'Unemployment Score']
data.head()
#data.to_csv('data/processed/unemp_crime.csv')
```

```python
data = pd.merge(whr_data, data, on=['Country Code', 'Year'])
data.head()
```

# Prosperity Index

```python
unemp_data = pd.read_csv('data/youth_unemployment.csv')
crime_data = pd.read_csv('data/crime_homicide.csv')

unemp_data = pd.melt(unemp_data, id_vars=['Country Name', 'Country Code', 'Indicator Name'], 
                     value_vars=['2000', '2001', '2002', '2003', '2004',
                   '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                   '2014', '2015', '2016', '2017', '2018', '2019', '2020'])

crime_data = pd.melt(crime_data, id_vars=['Country Name', 'Country Code', 'Indicator Name'], 
                     value_vars=['2000', '2001', '2002', '2003', '2004',
                   '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                   '2014', '2015', '2016', '2017', '2018', '2019', '2020'])

unemp_data.columns = ['Country Name', 'Country Code', 'Indicator Name', 'Year', 'Unemployment Score']
unemp_data['Year'] = pd.to_numeric(unemp_data['Year'], errors='coerce')
crime_data.columns = ['Country Name', 'Country Code', 'Indicator Name', 'Year', 'Crime (Homicide) Score']
crime_data['Year'] = pd.to_numeric(crime_data['Year'], errors='coerce')

data = pd.merge(crime_data, unemp_data, on=['Country Code', 'Year'])
```

```python
def clean_year(row):
    return row.split('_')[1]
```

```python
prosp_data = pd.read_excel('data/Legatum_Prosperity_Index_Full_2020_Data_Set.xlsx', engine='openpyxl')

prosp_data = pd.melt(prosp_data, id_vars=['area_name', 'area_code', 'area_group'], value_vars=[ 'score_2007', 'score_2008', 'score_2009',
       'score_2010', 'score_2011', 'score_2012', 'score_2013', 'score_2014',
       'score_2015', 'score_2016', 'score_2017', 'score_2018', 'score_2019',
       'score_2020'])

prosp_data['variable'] = prosp_data['variable'].apply(clean_year)
prosp_data.columns = ['Country', 'Country Code', 'Area Group', 'Year', 'Score']
prosp_data['Year'] = pd.to_numeric(prosp_data['Year'], errors='coerce')
```

```python
data = pd.merge(prosp_data, data, on=['Country Code', 'Year'])
data.head()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr())
```
